```python
#Q.1)	How many attacks happened in India?
terror[terror.country_txt=="India"].country_txt.count()
```

```python
#Q.4)In a single terror incident in India, find out top 5 cities by number killed
terror[(terror.nperps==1)&(terror.country_txt=="India")].sort_values("nkill",ascending=False)["city"].head(5)
```

```python
#Q.5)In a single terror incident in India, find out top 5 cities by number killed and wounded 
terror[(terror.country_txt=='India')].sort_values(by="nwound",ascending=False)[['city','nkill','nwound']].head(5)
```

```python
#Q.6)How many attacks were successful that were suicide attacks?
terror[(terror.success==1)&(terror.suicide==1)].success.count()
```

```python
#Q.7)label all the incidents where the number killed was more than 5 as severe.
terror['severity']=['severe' if i>5 else 'not severe'  for i in terror.nkill]
terror['severity']
```

```python
#Q.8)write a function to label an incident that was both successful and suicidal
def fun(dataset):
    j=["Success" if (dataset.iloc[i].suicide==1 and dataset.iloc[i].success==1) else "Failure" for i in list(range(0,len(dataset)))]
    return j

fun(terror)
```

```python
#Q.9)	Create a new category representing if the incident occured in Afghanistan, 
#Pakistan or India as one level of the category and all the other countries as another level
terror["category_new"]=np.nan
l=["Afghanistan","Pakistan","India"]
terror.loc[terror.country_txt.isin(l),"category_new"]="Asia"
terror.loc[~(terror.country_txt.isin(l)),"category_new"]="Rest_of_world"
terror["category_new"]
```

```python
#Q.10)	How many incidents happened in Af-Pak-India vs ROW?
terror["category_new"].value_counts()
```

```python
#Q.11)	List the number of suicides attacks and average kills by 
#Af-Pak-India vs ROW. Rename columns in the output as Average_Kills and Number_Incidents.
pd.crosstab(terror.category_new,terror.suicide,colnames=["Number_Incidents"])
df=terror.groupby("category_new")["nkill"].mean().reset_index()
df.columns=["Category","Average_Kills"]
df
```
